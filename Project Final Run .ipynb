# SVM test and demo

## Load pretrained SVM model and TF-IDF transformer

```python
class SVM_Classifier():
    def __init__(self, model_path = "./model_checkpoint/svm/svm.pkl",vectorizer_path = './model_checkpoint/svm/vectorzier.pkl'):
        with open(model_path, 'rb') as model:
            self.model = pickle.load(model)
        
        with open(vectorizer_path, 'rb') as vectorizer:
            self.vectorizer = pickle.load(vectorizer)
    
    # This method is used to predict the class of sentences
    def predict(self, sentence_list, true_class_name = False):
        if isinstance(sentence_list, str):
            sentence_list = [sentence_list]
        vectorized =  self.vectorizer.transform(sentence_list)
        predictions = self.model.predict(vectorized)
        
        class_names = {0: 'Negative', 1:'Neutral', 2:'Positive'}
        
        if true_class_name:
            class_predictions = []
            for num in predictions:
                class_name = class_names[num]
                class_predictions.append(class_name)
            
            final_predictions = class_predictions
        else:
            final_predictions = predictions
        
        return final_predictions

    # This method is used to predict the class probabilities of sentences
    def predict_proba(self, sentence_list):
        if isinstance(sentence_list, str):
            sentence_list = [sentence_list]

        vectorized =  self.vectorizer.transform(sentence_list)
        probabilities = self.model.predict_proba(vectorized)
        return probabilities

        
#         print(vectorized)
#         labels = ['Negative','Netural','Positive']
        
```

## Build vocabulary from the training set

```python
import collections
import os
import json
import datetime

class Vocabulary():
  '''Vocabulary class: extract all words from corpus, save all words that appears more than freq_threshold times'''
  def __init__(self, freq_threshold=3):
    self.itos = {0: "<PAD>", 1: "<SOS>", 2: "<EOS>", 3: "<UNK>"}
    self.stoi = {"<PAD>": 0, "<SOS>": 1, "<EOS>": 2, "<UNK>": 3}
    self.freq_threshold = freq_threshold

  def __len__(self):
    return len(self.itos)

  def build_vocab(self, sentence_list):
    # Build the vocab
    idx = 4
    word_counter = collections.Counter([word for sentence in sentence_list for word in sentence])

    for word, count in word_counter.items():
      if count >= self.freq_threshold:
        self.stoi[word] = idx
        self.itos[idx] = word
        idx += 1

  def numericalize(self, sent):
    # Turn a sentence into list of word ID
    return [
            self.stoi[token] if token in self.stoi else self.stoi["<UNK>"]
            for token in sent
        ]
  def save_vocab(self, vocab_path):
    with open(vocab_path, 'w') as f:
        vocab = {'itos': self.itos, 'stoi': self.stoi}
        json.dump(vocab, f)

  def load_vocab(self, vocab_path):
    with open(vocab_path, 'r') as f:
        vocab = json.load(f)
        self.itos = vocab['itos']
        self.stoi = vocab['stoi']
```

## Declare PyTorch dataset and dataloader

```python
import torch
import numpy as np
from torch.utils.data import DataLoader, Dataset
import random
import torch.nn.functional as F


seed  = 42
torch.manual_seed(seed)
torch.cuda.manual_seed_all(seed)
torch.cuda.manual_seed(seed)
np.random.seed(seed)
random.seed(seed)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

class SentimentDataset(Dataset):
  def __init__(self,df, vocab = None, max_sent_len = 120):
    self.train_sents = [sent.split(" ") for sent in df['sentences'].values]
    self.labels = df['sentiments'].values
    self.max_sent_len = max_sent_len
    if vocab == None:
        self.vocab = Vocabulary(3)
        self.vocab.build_vocab(self.train_sents)
    else:
        self.vocab = vocab
  def __len__(self):
    return len(self.train_sents)
      
  def __getitem__(self, index):
    sent_numericalized = [self.vocab.stoi["<SOS>"]]
    sent_numericalized += self.vocab.numericalize(self.train_sents[index])
    
    if len(sent_numericalized) > self.max_sent_len:
        sent_numericalized = sent_numericalized[:self.max_sent_len]

    sent_tensor = np.full(self.max_sent_len, self.vocab.stoi["<PAD>"])
    sent_tensor[(self.max_sent_len - len(sent_numericalized)):] = sent_numericalized
    x = torch.Tensor(sent_tensor).long()
    
    y = self.labels[index]
    return x,y
```

```python
def get_loader(df, vocab = None, max_sent_len = 120,batch_size = 64, num_workers = 0, shuffle = True, pin_memory = True):
    dataset = SentimentDataset(df, vocab = vocab, max_sent_len = max_sent_len)
    vocab = dataset.vocab
    loader = DataLoader(
        dataset=dataset,
        batch_size=batch_size,
        num_workers=num_workers,
        shuffle=shuffle,
        pin_memory=pin_memory
    )
    return loader, dataset, vocab
```

## Declare our LSTM model class

```python
class NetLSTM(nn.Module):
 
  def __init__(self, vocab_size, embedding_size, hidden_size, num_classes, num_layers, pad_idx, dropout, device, bidirectional):
    super(NetLSTM, self).__init__()
    self.vocab_size = vocab_size 
    self.embedding_size = embedding_size 
    self.hidden_size = hidden_size
    self.num_layers = num_layers
    self.device = device
    self.pad_idx = pad_idx
    self.num_classes = num_classes
    self.bidirectional = bidirectional

    self.dropout = nn.Dropout(dropout)
    
    #Embedding layer
    self.embedding = nn.Embedding(vocab_size, embedding_size, padding_idx = pad_idx).to(device)

    #LSTM layer
    self.lstm = nn.LSTM(
        input_size= self.embedding_size,
        hidden_size= self.hidden_size,
        batch_first=True,
        num_layers= self.num_layers
    ).to(device)

    #Linear, ReLU and Softmax layers
    self.fc = nn.Linear(in_features=self.hidden_size, out_features = self.num_classes).to(device)
    self.relu = nn.ReLU().to(device)
    self.softmax = nn.Softmax(dim = 1).to(device)

  def forward(self, x):
    h_0 = torch.zeros(self.num_layers, x.shape[0], self.hidden_size).to(self.device) #hidden state
    c_0 = torch.zeros(self.num_layers, x.shape[0], self.hidden_size).to(self.device) #internal state

    #perform dropout in embedding layer
    embedded = self.dropout(self.embedding(x))

    #LSTM layer
    output, (hn, cn) = self.lstm(embedded, (h_0, c_0)) 
   
    #reshaping the data for Dense layer
    hn = hn[-1].view(-1, self.hidden_size) 

    #fully connected layer and activation layers (ReLU and Softmax)
    out = self.fc(hn)  
    out = self.relu(out) 
    out = self.softmax(out) #Final Output

    return out
```

## Declare our CNN model class

```python
class NetCNN(nn.Module):
    def __init__(self, vocab_size, embed_size, kernel_sizes, num_channels, drop, pad_idx):
        super(NetCNN, self).__init__()
        self.pad_idx = pad_idx
        self.embedding = nn.Embedding(vocab_size, embed_size, padding_idx = pad_idx)
        
        # The embedding layer not to be trained
        self.constant_embedding = nn.Embedding(vocab_size, embed_size, padding_idx = pad_idx)
        self.dropout = nn.Dropout(drop)
        self.fc = nn.Linear(in_features = sum(num_channels), out_features = 3)

        # The max-over-time pooling layer has no parameters, so this instance
        # can be shared
        self.pool = nn.AdaptiveAvgPool1d(1)
        self.relu = nn.ReLU()

        # Create multiple one-dimensional convolutional layers
        self.convs = nn.ModuleList()
        for c, k in zip(num_channels, kernel_sizes):
            self.convs.append(nn.Conv1d(2 * embed_size, c, k))

    def forward(self, inputs):
        # Concatenate two embedding layer outputs with shape (batch size, no.
        # of tokens, token vector dimension) along vectors
        embeddings = torch.cat((self.embedding(inputs), self.constant_embedding(inputs)), dim=2)

        # Per the input format of one-dimensional convolutional layers,
        # rearrange the tensor so that the second dimension stores channels
        embeddings = embeddings.permute(0, 2, 1)

        # For each one-dimensional convolutional layer, after max-over-time
        # pooling, a tensor of shape (batch size, no. of channels, 1) is
        # obtained. Remove the last dimension and concatenate along channels
        conv = torch.cat([
            torch.squeeze(self.relu(self.pool(conv(embeddings))), dim=-1)
            for conv in self.convs], dim=1)
        output = self.fc(self.dropout(conv))
        return output
```

## Declare our NN model class

```python
class NetLinear(nn.Module):
 
  def __init__(self, vocab_size, embedding_size, hidden_size, num_classes, num_layers, pad_idx, dropout, device):
    super(NetLinear, self).__init__()
    self.vocab_size = vocab_size 
    self.embedding_size = embedding_size 
    self.hidden_size = hidden_size
    self.num_layers = num_layers
    self.device = device
    self.pad_idx = pad_idx
    self.num_classes = num_classes

    self.dropout = nn.Dropout(dropout)
    
    #Embedding layer
    self.embedding = nn.Embedding(vocab_size, embedding_size, padding_idx = pad_idx).to(device)

    #Linear, ReLU and Softmax layers
    self.fc = nn.Linear(in_features=self.embedding_size * 100, out_features = self.num_classes).to(device)
    self.relu = nn.ReLU().to(device)
    self.softmax = nn.Softmax(dim = 1).to(device)

  def forward(self, x):
    embedded = self.embedding(x)
    embedded = embedded.view(-1, self.embedding_size*100)
    
    out = self.dropout(self.fc(embedded))
    out = self.relu(out) 
    out = self.softmax(out) #Final Output

    return out
```

## Create LSTM model

```python
vocab_size = len(vocab)
embedding_size = 256
hidden_size = 128
num_classes = 3
num_layers = 1
dropout = 0.5
bidirectional = False
pad_idx = vocab.stoi["<PAD>"]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu") # train on GPU if available


model_lstm = NetLSTM(vocab_size = vocab_size, embedding_size = embedding_size, hidden_size = hidden_size, num_classes = num_classes, num_layers = num_layers, dropout = dropout, pad_idx = pad_idx, device = device, bidirectional = bidirectional)
```

## Create CNN model

```python
PAD_IDX = vocab.stoi["<PAD>"]
INPUT_DIM = len(vocab)
EMBEDDING_DIM = 256
N_FILTERS = [100, 100, 100]
FILTER_SIZES = [3, 4, 5]
OUTPUT_DIM = 3
DROPOUT = 0.5

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model_cnn = NetCNN(INPUT_DIM, EMBEDDING_DIM, FILTER_SIZES, N_FILTERS, DROPOUT, PAD_IDX).to(device)
```

## Create Linear model

```python
vocab_size = len(vocab)
embedding_size = 256
hidden_size = 128
num_classes = 3
num_layers = 2
dropout = 0.2
pad_idx = vocab.stoi["<PAD>"]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu") # train on GPU if available


model_linear = NetLinear(vocab_size = vocab_size, embedding_size = embedding_size, hidden_size = hidden_size, num_classes = num_classes, num_layers = num_layers, dropout = dropout, pad_idx = pad_idx, device = device)
```
